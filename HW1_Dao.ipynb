```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import uncertainties as unc
from scipy.stats import norm 
import scipy.constants as const
%matplotlib inline
```

# Problem 1
Generate 100k data points from a gaussian distribution with mean $\mu=3$ and width $\sigma=0.01$

```python
# Generate a set of random data points, d, from requirements above
d = norm.rvs(loc = 3.0, scale = 0.01, size = 100000)
```

# Problem 2
Plot a histogram of these points using matplotlib. Set x-axis range to $2.95-3.05$ and the number of bins to 50.

```python
# Set up fig and plot histogram 
fig, ax = plt.subplots(1, 1)
ax.hist(d, 50, density=True)
plt.xlim([2.95, 3.05])
```

# Problem 3
Superimpose a Gaussian probability density function with normalization to 100k.

```python
# Generate array of equally spaced x-values and plot a gaussian probability density function with them by applyling norm.pdf
fig, ax = plt.subplots(1, 1)
ax.hist(d, 50, density=True)
plt.xlim([2.95, 3.05])
x = np.linspace(2.95, 3.05, 1000)
ax.plot(x, norm.pdf(x, loc=3.0, scale=0.01), linewidth=5, alpha=0.6)
plt.show()
```

# Problem 4
Redraw the same histogram and PDF using log scale on the y-axis.

```python
# Copy from problem 3 and adjust y-axis scale using pyplot yscale
fig, ax = plt.subplots(1, 1)
ax.hist(d, 50, density=True)
plt.yscale('log')
x = np.linspace(2.95, 3.05, 1000)
ax.plot(x, norm.pdf(x, loc=3.0, scale=0.01), linewidth=5, alpha=0.6)
plt.xlim([2.95, 3.05])
plt.show()
```
